```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python


# Replace this path with your actual Excel file path
data_path1 = "/home/likith/DBMS/outputs/result_1.csv/part-00000-8b3e9d3a-b40b-478e-b490-8ffe7785fb3c-c000.csv"

# Read data from Excel sheet
df = pd.read_csv(data_path1)


```

```python
# analysis 1 
# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['avg'], labels=df['Segment'], autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'orange'])
plt.title('Average Values by Segment (Pie Chart)')
plt.show()
```

```python

# Replace this path with your actual Excel file path
data_path2 = "/home/likith/DBMS/outputs/result_2.csv/part-00000-f08a43a9-c93b-4c71-af6c-3bfb73a0d285-c000.csv"

# Read data from Excel sheet
df2 = pd.read_csv(data_path2)
```

```python
# analysis2
# Pivot the data for better plotting
pivot_df = df2.pivot(index='Month', columns='Sub_Category', values='TotalSales')

# Set seaborn style
sns.set(style="whitegrid")

# Plotting with seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_df, markers=True, dashes=False)

plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales Trend by Sub_Category')
plt.legend(title='Sub_Category')
plt.show()
```

```python
# analysis6
data_path5 = "/home/likith/DBMS/outputs/result_6.csv/part-00000-6a84051f-62bc-4609-8249-da5d5dfca43f-c000.csv"

# Read data from Excel sheet
df5 = pd.read_csv(data_path5)
```

```python
# Scatter plot for total profit vs. rank
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Rank', y='TotalProfit', data=df5, hue='Year', palette='viridis')
plt.xlabel('Rank')
plt.ylabel('Total Profit')
plt.title('Total Profit vs. Rank')
plt.tight_layout()
plt.show()
```

```python
# Set up subplots
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(15, 15))
fig.suptitle('Total Profit Over the Years for Top Products', fontsize=16)

# Iterate through each rank and create a line plot
for rank, ax in zip(range(1, 6), axes.flatten()):
    top_product = df5[df5['Rank'] == rank]
    
    sns.lineplot(x='Year', y='TotalProfit', data=top_product, marker='o', color='blue', ax=ax)
    ax.set_title(f'Rank {rank}')
    ax.set_xlabel('Year')
    ax.set_ylabel('Total Profit')

# Adjust layout and show the plot
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
